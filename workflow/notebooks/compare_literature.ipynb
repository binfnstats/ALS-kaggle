```python
import pandas as pd
```

```python
import os

try:
    snakemake
except NameError:
    from snakemk_util import load_rule_args
    snakemake = load_rule_args(
        snakefile = os.getcwd() + '/../Snakefile',
        rule_name = 'compare_literature',
        root=os.getcwd() + "/.."
)
```

```python
df_results = pd.read_csv(snakemake.input['results']).set_index('geneID')
df_genes = pd.read_csv(snakemake.input['genes']).set_index('geneID')
df_nature = pd.read_csv(snakemake.input['nature_review'])
df_wgs = pd.read_csv(snakemake.input['wgs'])
```

```python
df_results = df_results.join(df_genes)
```

```python
genes_nature = set(df_results['gene_name']).intersection(df_nature['gene_name'])

with open(snakemake.output['nature_review'], 'w') as f:
    for i in genes_nature:
        f.write(i + '\n')

genes_nature
```

```python
genes_wgs = set(df_results['gene_name']).intersection(df_wgs['gene_name'])

with open(snakemake.output['wgs'], 'w') as f:
    for i in genes_wgs:
        f.write(i + '\n')

genes_wgs
```
